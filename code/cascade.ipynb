# Image (OpenCV)

```python
img = cv2.imread(imagePath)
img.shape
```

```python
img = cv2.resize(img, (img.shape[1]//2, img.shape[0]//2))
img.shape
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape
```

# Face Cascade Classifier

```python
face_classifier  = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
eyes_classifier  = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
smile_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")
#upper_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_upperbody.xml")
#body_classifier  = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_fullbody.xml")
```

```python
face = face_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=1, minSize=(10, 10))
#eyes = eyes_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=3, minSize=(2, 2))
#smil = smile_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=20, minSize=(5, 2))
#uppb = upper_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=10, minSize=(30, 50))
#body = body_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=10, minSize=(30, 100))
```

```python
face.shape
```

```python
face
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```
